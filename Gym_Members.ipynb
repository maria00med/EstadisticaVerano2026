## **Introducción**

La cadena de gimnasios FitLife busca mejorar la experiencia de sus usuarios mediante la personalización científica de sus rutinas. El problema central radica en la necesidad de predecir de forma precisa el gasto calórico de los socios. La gerencia hipotetiza que la duración de la sesión es el factor determinante y desea comprobar si existe una relación lineal lo suficientemente sólida como para estandarizarla como regla de negocio automatizada.



Para este reporte preliminar, se analizan exhaustivamente dos métricas clave de las sesiones:

Variable Predictora: Duración de la sesión de entrenamiento (registrada en horas).

Variable de Interés: Calorías quemadas totales (medidas en kilocalorías, kcal).


Comprender este vínculo matemático permite a los entrenadores e instructores de FitLife diseñar rutinas con objetivos garantizados. En lugar de estimar de manera intuitiva, el gimnasio podrá ofrecer certezas a los socios (por ejemplo: "Si entrenas 45 minutos en esta intensidad, tu gasto programado será de X calorías"), incrementando la retención, la satisfacción y la motivación del cliente.



Evaluar la viabilidad, fuerza y confiabilidad de un modelo matemático basado en la duración de la sesión para predecir el gasto calórico de los socios, determinando si una sola variable es suficiente para la planeación operativa o si se requiere un enfoque multifactorial.

## **Metodología**

Se utilizó el Gym Members Exercise Dataset, una base de datos institucional que recopila información detallada de 973 sesiones de entrenamiento individuales de socios con perfiles heterogéneos en cuanto a edad, peso, género y niveles de condición física.


Se detectaron las siguientes variables:

Variable Independiente ($X$): Duración de la sesión (en horas). Es la variable que el gimnasio puede manipular y programar en la rutina.

Variable Dependiente ($Y$): Calorías quemadas ($kcal$). Es la respuesta metabólica resultante que deseamos estimar.


Se aplicó la técnica de Regresión Lineal Simple por Mínimos Cuadrados Ordinarios (OLS). Se eligió este enfoque porque la gerencia busca una herramienta operativa ágil, una "regla de dedo" matemática que sea fácil de interpretar por el personal en piso de entrenamiento y fácil de programar en la aplicación móvil de FitLife, sin la complejidad de requerir software estadístico avanzado en el día a día.


Para dar validez científica al informe se emplearon:

* Gráfico de Dispersión (Scatter Plot): Para examinar visualmente la tendencia, agrupación y posible presencia de anomalías en los datos.

* Coeficiente de Correlación de Pearson ($r$): Para medir la fuerza y dirección de la asociación entre el tiempo y las calorías.

* Coeficiente de Determinación ($R^2$): Para conocer qué porcentaje de la variabilidad calórica es explicada por el tiempo entrenado.

* Recta de Regresión Ajustada: Es la representación visual y matemática de la tendencia o relación promedio que existe entre dos variables (una independiente $X$ y una dependiente $Y$).

* Análisis Gráfico de Residuales: Para auditar los errores de predicción y verificar si el modelo es estable a lo largo de todas las duraciones de rutina.

* Predicciones: Permiten calcular con un alto grado de certeza qué pasará con la variable dependiente ante un valor específico de la variable independiente, incluso si nunca antes se ha registrado ese valor exacto.

## **Resultados**

El gráfico de dispersión muestra una tendencia lineal ascendente clara y bien definida. A medida que aumenta el valor en el eje horizontal (Duración de la sesión en horas), se observa un incremento consistente y proporcional en el eje vertical (Calorías quemadas). Los puntos de los 973 socios se agrupan de forma compacta alrededor de la trayectoria de una línea recta, lo que visualmente descarta comportamientos curvilíneos o una dispersión caótica de los datos.

El coeficiente de correlación de Pearson (r) obtenido es de 0.9081. Este valor se interpreta como una correlación positiva muy fuerte. En el contexto del gimnasio, demuestra que existe una asociación directa y altamente confiable entre ambas métricas: a mayor tiempo de entrenamiento, el gasto calórico se incrementa de forma predecible.

El coeficiente de determinación ($R\text{-squared}$) es de 0.8247. Esto significa que el 82.47% de la variabilidad total en las calorías quemadas por los socios es explicada única y directamente por la duración de la sesión. El restante de la variación se debe a variables individuales externas que no forman parte de este modelo simple (como el peso, la edad o la intensidad del ejercicio).

A partir de los coeficientes estimados por la herramienta estadística, la ecuación matemática oficial para el gimnasio es: Calorías Quemadas = 3.18 + 717.38 * (Duración en Horas). La pendiente del modelo es de 717.38. En el contexto operativo de FitLife, esto significa que por cada hora adicional que un socio permanezca entrenando, su gasto calórico se incrementará en promedio 717.38 kcal. Expresado de forma más práctica para el personal en piso, cada 15 minutos de entrenamiento representan un aproximado de 179.3 kcal quemadas.

El gráfico de residuales muestra una nube de puntos distribuida de manera mayormente homogénea y aleatoria alrededor de la línea horizontal del cero. No se aprecian patrones geométricos marcados ni formas de "embudo" críticas (heterocedasticidad).

Esto indica que los errores de predicción del modelo son estables y no presentan sesgos sistemáticos graves, confirmando que la calidad y los supuestos de la regresión lineal se cumplen adecuadamente para la planeación comercial.

Tomando como referencia el escenario planteado para la toma de decisiones, si un entrenador configura una rutina personalizada con una duración de 1.5 horas (90 minutos), el modelo matemático realiza la siguiente predicción: Calorías = 3.18 + 717.38 * (1.5) = 3.18 + 1076.07 = 1,079.25 kcal.

El sistema arroja de forma automatizada un gasto calórico estimado de 1,079.25 kilocalorías para esa sesión.

### Interpretación del Gráfico de Dispersión

El gráfico de dispersión muestra la distribución de los puntos de datos, donde cada punto representa una sesión de entrenamiento con su duración y las calorías quemadas correspondientes. Esto nos permite identificar visualmente:

*   **Tendencia:** Si los puntos tienden a seguir un patrón ascendente (relación positiva), descendente (relación negativa) o si no hay un patrón claro.
*   **Fuerza de la relación:** Qué tan agrupados están los puntos alrededor de una posible línea recta.
*   **Valores atípicos (Outliers):** Puntos que se desvían significativamente del patrón general.

Para cuantificar esta relación de manera más precisa, procedemos a calcular el Coeficiente de Correlación de Pearson y el Coeficiente de Determinación (R²).

### Coeficiente de Correlación de Pearson ($r$)

```python
import pandas as pd
from scipy.stats import pearsonr

# The DataFrame 'df' and variables 'X' and 'Y' are expected to be defined
# globally from cell 'IAph_oT20cCD'.
# The previous lines attempting to redefine df, X, and Y were causing errors
# and using incorrect column names.

# Test de Pearson
# H0: rho = 0     (No hay correlación)
# H1: rho ≠ 0     (Sí hay correlación)
# alpha = 0.05

# Ensure X and Y are defined and not empty before calculating Pearson's r
if 'X' in locals() and 'Y' in locals() and not X.empty and not Y.empty:
    r, valor_p = pearsonr(X, Y)

    print(f"Coeficiente de correlación: {r: 0.4f}")
    print(f'valor_p: {valor_p: 0.4f}')
else:
    print("Variables X and/or Y are empty or not defined. Please ensure data is loaded and variables are set correctly from cell 'IAph_oT20cCD'.")
```

### Coeficiente de Determinación ($R^2$)

```python
import statsmodels.api as sm
# Coeficiente de determinación
from sklearn.metrics import r2_score

x_constante = sm.add_constant(X)
modelo = sm.OLS(Y, x_constante).fit()
y_calculada = modelo.predict(x_constante)

r2 = r2_score(Y, y_calculada)

print(f"Coeficiente de determinación: {r2: 0.2%}")
```

En otras palabras, nos dice qué tan bien el modelo de regresión explica la variabilidad de las calorías quemadas.
