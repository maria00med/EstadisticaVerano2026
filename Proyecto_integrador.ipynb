## **1. Introducción**

El presente proyecto integrador aborda un problema tecnológico y económico de alta relevancia en la sociedad actual: la cuantificación y el modelado estadístico del impacto del tiempo de uso, medido en meses de funcionamiento continuo, sobre la degradación de la capacidad máxima de retención de carga en las baterías de dispositivos inteligentes de la marca Apple (iPhone). Las baterías de iones de litio sufren un desgaste químico natural e inevitable con el paso del tiempo. Para los usuarios cotidianos, los técnicos de soporte y el creciente mercado de dispositivos seminuevos, comprender con precisión la tasa real de este deterioro resulta fundamental para estimar la vida útil restante del equipo, planificar de manera financiera un reemplazo preventivo y establecer un valor comercial justo para los dispositivos usados que esté fundamentado en evidencia estadística y no en meras especulaciones subjetivas.

Ante este escenario, la expectativa inicial y la hipótesis de investigación planteada por el equipo postulan que existe una relación lineal inversa, de carácter negativo y estadísticamente significativa, entre los meses de uso acumulados por el dispositivo y su porcentaje de salud de batería. En términos prácticos, se espera que a mayor cantidad de meses de funcionamiento, la capacidad máxima de retención disminuya de forma proporcional y constante. Por lo tanto, el objetivo principal de este proyecto consiste en desarrollar y evaluar un modelo de regresión lineal simple que explique y prediga dicha capacidad en función del tiempo de activación, evaluando de forma rigurosa la calidad del ajuste matemático y la validez de los supuestos estadísticos correspondientes mediante los datos reales obtenidos directamente por el equipo.



---



## **2. Metodología**

Para la realización de este estudio se optó por un método de recopilación directa y primaria mediante la observación y el muestreo físico de campo, asegurando la autenticidad de la información y cumpliendo estrictamente con la restricción de no emplear bases de datos prefabricadas por terceros. Cada integrante del equipo consultó de forma presencial o remota el apartado nativo de "Salud de la Batería" en los ajustes del sistema operativo iOS de diversos dispositivos iPhone pertenecientes a amigos, familiares y de uso propio. La muestra final recolectada, que representa a la población de teléfonos celulares iPhone activos y bajo condiciones de uso cotidiano real en el entorno urbano de los usuarios, consta de un tamaño de muestra de exactamente n=21 observaciones independientes.

En este diseño experimental, la variable independiente (X) quedó definida como los meses de funcionamiento continuo acumulados desde que el dispositivo fue activado por primera vez, tratándose de una variable cuantitativa continua medida en una escala de razón que se aproximó a meses enteros a través del tiempo transcurrido desde la fecha de compra o activación inicial. Por su parte, la variable dependiente (Y) se definió como la capacidad máxima de la batería, una variable cuantitativa continua expresada en porcentaje dentro de un rango observado de 71% a 100%, la cual refleja la salud y retención actual del componente en comparación con su estado original de fábrica. Ambas variables se midieron de manera digital y directa a través de la interfaz oficial de diagnóstico del dispositivo, minimizando errores de lectura manual.

El procedimiento estadístico aplicado consistió en el análisis de correlación de Pearson y la estimación de parámetros a través del método de Mínimos Cuadrados Ordinarios (MCO). La regresión lineal simple se determinó como el procedimiento adecuado para este problema dado que la investigación plantea analizar la interacción y dependencia entre una sola variable predictora cuantitativa y una sola variable de respuesta continua, alineándose además con la teoría de degradación de componentes electrónicos que sugiere que el desgaste químico de las celdas de litio mantiene una tasa de decremento aproximadamente constante y proporcional a lo largo del mediano plazo.

## **3. Resultados**

A partir del análisis de los datos recolectados, el gráfico de dispersión (que cruza los meses de funcionamiento con la capacidad máxima de la batería) muestra una clara tendencia descendente y decreciente. A medida que avanzan los meses en el eje horizontal, se observa visualmente cómo los puntos de los dispositivos móviles se desplazan hacia valores inferiores en la salud de la batería, lo que representa una evidencia clara y visual de la existencia de una relación lineal negativa.

Esta tendencia se confirma matemáticamente a través de las métricas obtenidas del modelo. El coeficiente de correlación de Pearson (r) se situó en -0.6018, lo que se interpreta formalmente como una asociación lineal inversa de fuerza moderada a buena, ratificando con rigor estadístico que un mayor tiempo de uso se vincula de forma directa con la pérdida de capacidad de carga. Asimismo, el coeficiente de determinación (R
^2
) resultó ser de 0.3621, indicando que el modelo de regresión lineal explica el 36.21% de la variabilidad total observada en la capacidad máxima de la batería, lo que deja el 63.79% restante a la influencia de variables externas que no fueron medidas en este estudio inicial.

A través del método de mínimos cuadrados ordinarios, la ecuación de regresión obtenida para este proyecto representa la capacidad máxima estimada y X el tiempo en meses. En el contexto de este proyecto, la pendiente de -0.3275 es el indicador clave de desgaste, significando que por cada mes adicional de funcionamiento continuo que acumula un iPhone, su capacidad máxima disminuye en promedio un 0.3275%, lo que se traduce en un desgaste estimado de 3.93% por cada año transcurrido.

El análisis de los residuales —las diferencias entre los valores reales observados y las predicciones matemáticas del modelo— revela una dispersión aleatoria y homogénea a ambos lados de la línea de cero. Al no observarse patrones curvos, embudos o acumulaciones asimétricas de los puntos, los residuales indican que el modelo posee una calidad aceptable y que se cumplen de forma satisfactoria los supuestos fundamentales de linealidad y homocedasticidad. Esto confirma que la variabilidad del error se mantiene constante y que el modelo conserva su misma precisión predictiva tanto para dispositivos nuevos como para aquellos con un tiempo de uso avanzado. Finalmente, para demostrar el alcance analítico de la ecuación obtenida, se puede realizar una predicción práctica: para un dispositivo iPhone que alcance exactamente los 24 meses (2 años) de funcionamiento continuo, el modelo estima que su capacidad máxima de batería se situará en un 83.81%.

```python
# Importamos la librería matplotlib.pyplot y le damos el pseudónimo plt
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(
    {
        "Meses_De_Funcionamiento":[12,6,51,15,16,34,33,48,36,35,5,6,7,59,26,20,30,15,4,24,13],
        "Capacidad_Máxima_(%)":[82,76,81,87,79,84,80,79,71,76,100,100,97,73,87,78,84,80,100,89,80]
    }
)

X= df["Meses_De_Funcionamiento"]
Y= df["Capacidad_Máxima_(%)"]

# Configuración general del gráfico
plt.figure(
    figsize=(6, 4),   # tamaño de la figura (ancho, alto) en pulgadas
    dpi=120           # resolución del gráfico
)

# Gráfico de dispersión
plt.scatter(
    X, Y,
    marker="o",       # forma: googlear "matplotlib.markers"
    color='yellow',     # color de los puntos
    edgecolor='yellow',    # borde de los puntos
    alpha=0.9,            # transparencia: 0 es invisible y 1 es color sólido
    s=50                 # tamaño de los puntos
)

# Etiquetas de los ejes
# eje x
plt.xlabel(
    'Meses de Funcionamiento', # etiqueta del eje x
    fontsize=12  # tamaño de fuente
)

plt.ylabel(
    'Capacidad Máxima (%)', # etiqueta del eje y
    fontsize=12 # tamaño de fuente
)

# Fuente de los ticks
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Márgenes
plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior

# Cuadrícula (opcional, pero didáctica)
plt.grid(
    visible=True,
    linestyle='--',
    linewidth=0.7,
    alpha=0.5,
    color="gray"
)
```

```python
import pandas as pd
from scipy.stats import pearsonr

df = pd.DataFrame(
{"Meses_De_Funcionamiento":
 [12,6,51,15,16,34,33,48,36,35,5,6,7,59,26,20,30,15,4,24,13],
        "Capacidad_Máxima_(%)":
         [82,76,81,87,79,84,80,79,71,76,100,100,97,73,87,78,84,80,100,89,80]
            }
            )



X= df["Meses_De_Funcionamiento"]
Y= df["Capacidad_Máxima_(%)"] # Corregido el nombre de la columna

# Test de Pearson
# H0: rho = 0     (No hay correlación)
# H1: rho ≠ 0     (Sí hay correlación)
# alpha = 0.05

r, valor_p = pearsonr(X, Y)

print(f"Coeficiente de correlación: {r: 0.4f}")
print(f'valor_p: {valor_p: 0.4f}')
```

```python
import statsmodels.api as sm
x_constante = sm.add_constant(X)
modelo = sm.OLS(Y, x_constante).fit()
y_calculadora = modelo.predict(x_constante)
```

```python
import statsmodels.api as sm
# Coeficiente de determinación
from sklearn.metrics import r2_score

x_constante = sm.add_constant(X)
modelo = sm.OLS(Y, x_constante).fit()
y_calculada = modelo.predict(x_constante)

r2 = r2_score(Y, y_calculada)

print(f"Coeficiente de determinación: {r2: 0.2%}")
```
